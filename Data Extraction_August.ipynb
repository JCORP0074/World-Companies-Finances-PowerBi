```python
import pandas as pd
import numpy as np 
from sqlalchemy import create_engine
import glob
import os
```

<p style="font-size:20px; text-align:justify;">
  The Top Companies Dataset is categorized according to industry types. It includes the following columns: 
  <span style="font-weight:bold;">Name, Symbol, Country, Industry, Number of Employees , Price (USD), Revenue,</span> 
  <em>and</em> 
  <span style="font-weight:bold;">Earnings.</span> 
  The dataset reflects information as of August 2025.
</p>


```python
def get_data_categorize(datafile,category):
    filepath = glob.glob(datafile)
    maindata = pd.read_csv(filepath[0])
    for file in filepath[1:]:
        second_data = pd.read_csv(file)
        maindata = pd.merge(maindata, second_data, on="Name", how="inner")
        maindata["Industry"] = category
    columns_needed = ["Name","Symbol","country","Industry","employees_count","price (USD)", "revenue_ttm","earnings_ttm"]
    maindata = maindata[columns_needed]
    os.makedirs("../Companies CSVs/", exist_ok=True)
    maindata.to_csv(f"../Companies CSVs/{category}_data.csv", index=False)
    return maindata
    
```

<p style = "font-size:40px; font-weight: bold"> UTILITY INDUSTRY DATA (ELECTRICITY, GAS, AND OIL)</p>

```python
def concat_df(df1,df2,industry):
    concated_data = pd.concat([df1,df2], ignore_index=True)
    concated_data["Industry"] = concated_data["Industry"].replace(
                                ["Electricity Industry", "Oil And Gas Industry"],
                                industry
)

    concated_data = concated_data.drop_duplicates(subset=["Name"], keep="first")
    return concated_data
```

```python
utility_data = concat_df(electricity_data,oil_and_gas,"Utilities")
```
